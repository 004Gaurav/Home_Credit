```python
# Outlier Detection Techniques:
# 1. Visual Methods:
#    - Boxplot: Outliers appear as points beyond the whiskers.
#    - Distribution Plot (distplot): Shows skewness and long tails indicating possible outliers.
#
# 2. Statistical Methods:
#    - Analyze the spread between percentiles:
#      • Check the difference between Q3 (75th percentile) and Q1 (25th percentile).
#      • Also observe the difference between the 75th and 100th percentile.
#      • If the value range between 75th and 100th percentile is significantly larger than 0–25% or 25–75%,
#        this may indicate outliers.
#
# Note:
# A "huge difference" generally means:
#   (100th percentile - 75th percentile) >> (75th - 25th) or (25th - 0th)
# This suggests values are stretching unusually far on the higher end — a possible sign of outliers.
```

```python
poly_features = app_train[['TARGET','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']]
poly_features_test = app_test[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']]
```

```python
from sklearn.impute import SimpleImputer
poly_target = poly_features['TARGET']
poly_features = poly_features.drop(columns = ['TARGET'])
```

```python
imputer = SimpleImputer(strategy='median')
poly_features_imputed = imputer.fit_transform(poly_features)
poly_features_test_imputed = imputer.fit_transform(poly_features_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_transformer = PolynomialFeatures(degree=3)
poly_transformer.fit(poly_features_imputed)
```

```python
poly_features = poly_transformer.transform(poly_features_imputed)
poly_features_test = poly_transformer.transform(poly_features_test_imputed)
```

```python
poly_features.shape
```

```python
poly_transformer.get_feature_names_out(input_features=['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH'])
```

```python
app_train.shape
```

```python
poly_features = pd.DataFrame(poly_features,
                             columns=poly_transformer.get_feature_names_out(
                                 ['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3','DAYS_BIRTH']
                             ))
poly_features['TARGET'] = poly_target.values

poly_corrs = poly_features.corr()['TARGET'].sort_values()
```

```python
poly_corrs
```

```python
# Put test features into dataframe with correct feature names
poly_features_test = pd.DataFrame(poly_features_test,
                                  columns=poly_transformer.get_feature_names_out(
                                      ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH']
                                  ))

# Add SK_ID_CURR to train and test polynomial feature sets
poly_features['SK_ID_CURR'] = app_train['SK_ID_CURR']
poly_features_test['SK_ID_CURR'] = app_test['SK_ID_CURR']

# Merge polynomial features into training and testing data
app_train_poly = app_train.merge(poly_features, on='SK_ID_CURR', how='left')
app_test_poly = app_test.merge(poly_features_test, on='SK_ID_CURR', how='left')

# Align the dataframes by keeping only the columns present in both
app_train_poly, app_test_poly = app_train_poly.align(app_test_poly, join='inner', axis=1)

# Output new shapes
print('Training data with polynomial features shape: ', app_train_poly.shape)
print('Testing data with polynomial features shape:  ', app_test_poly.shape)
```

```python
# Re-load the original dataframes to ensure 'DAYS_BIRTH' is present
app_train_original = pd.read_csv("application_train.csv")
app_test_original = pd.read_csv("application_test.csv")

# Perform one-hot encoding on the original dataframes
app_train_original = pd.get_dummies(app_train_original, dtype=int)
app_test_original = pd.get_dummies(app_test_original, dtype=int)

# Align the original dataframes after one-hot encoding
app_train_original, app_test_original = app_train_original.align(app_test_original, join='inner', axis=1)


# Create domain features using the original dataframes
app_train_domain = app_train_original.copy()
app_test_domain = app_test_original.copy()

app_train_domain['CREDIT_INCOME_PERCENT'] = app_train_domain['AMT_CREDIT'] / app_train_domain['AMT_INCOME_TOTAL']
app_train_domain['DAYS_EMPLOYED_PERCENT'] = app_train_domain['DAYS_EMPLOYED'] / app_train_domain['DAYS_BIRTH']

app_test_domain['CREDIT_INCOME_PERCENT'] = app_test_domain['AMT_CREDIT'] / app_test_domain['AMT_INCOME_TOTAL']
app_test_domain['DAYS_EMPLOYED_PERCENT'] = app_test_domain['DAYS_EMPLOYED'] / app_test_domain['DAYS_BIRTH']

# Merge the domain features back to the dataframes with polynomial features
# Assuming app_train and app_test at this point already contain the polynomial features
# and the target variable is in app_train
app_train = app_train_domain.merge(poly_features, on='SK_ID_CURR', how='left')
app_test = app_test_domain.merge(poly_features_test, on='SK_ID_CURR', how='left')


# Align the dataframes
app_train_poly, app_test_poly = app_train.align(app_test, join='inner', axis=1)


print('Training data with polynomial and domain features shape:', app_train_poly.shape)
print('Testing data with polynomial and domain features shape:', app_test_poly.shape)
```
